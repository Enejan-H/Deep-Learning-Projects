## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
train_scaled = scaler.fit_transform(train)
test_scaled = scaler.transform(test)
```

Bu kod paras, veri nileme admnda yer almaktadr. scaler adl bir lekleyici nesnesi kullanlarak, eitim ve test verileri leklendirilmektedir.

fit_transform() yntemi, scaler nesnesini eitim verilerine uygulayarak verileri leklendirir ve dntrr. Eitim verileri zerinde leklendirme ilemi yaplrken, lekleyici nesne eitim verilerine uyum salamak iin verileri analiz eder ve istatistiksel bilgileri kullanarak verileri leklendirir.

Daha sonra, transform() yntemi kullanlarak ayn lekleyici nesne, test verilerine uygulanr. Bu durumda, lekleyici nesne, eitim verilerine dayal olarak rendii lekleme faktrlerini kullanarak test verilerini leklendirir. Bu ekilde, eitim ve test verileri ayn leklendirme faktrlerine gre uygun bir ekilde dntrlr.

Bu ilem, verilerin leklendirilerek ayn arala getirilmesini salar. leklendirme, modelin daha iyi performans gstermesine ve daha istikrarl sonular retmesine yardmc olabilir.

# Modelling & Model Performance

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM, SimpleRNN
from keras.callbacks import EarlyStopping
from keras.layers import Dropout
```

```python
train_scaled.shape
```

```python
n_features = train_scaled.shape[1]
n_features
```

## with "tanh" activation function

```python
model = Sequential()

model.add(LSTM(512, activation = 'tanh', return_sequences=True, input_shape = (length, n_features)))

model.add(Dropout(0.2))  # Dropout katman eklendi, dropout oran 0.2 olarak ayarland

model.add(LSTM(256, activation = 'tanh', return_sequences=True))
model.add(Dropout(0.2))  # Dropout katman eklendi, dropout oran 0.2 olarak ayarland

model.add(LSTM(128, activation = 'tanh', return_sequences=True))
model.add(Dropout(0.2)) 

model.add(LSTM(64, activation = 'tanh', return_sequences=True))
model.add(Dropout(0.2)) 

model.add(LSTM(16, activation = 'tanh'))
model.add(Dropout(0.2))  # Dropout katman eklendi, dropout oran 0.2 olarak ayarland

model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse')
```

LSTM (Uzun Ksa-Termli Hafza) katman, zaman serileri veya ardk veriler gibi sral veri yaplarnn analizinde kullanlan bir tr yapay sinir a katmandr. LSTM'nin en nemli zelliklerinden biri, gizli hcrelerinde gemi bilgileri koruyabilmesidir.

Bir LSTM katman, genellikle ardk zaman admlarnn ilenmesi iin kullanlr. rnein, bir zaman serisi tahminlemesi yaparken, LSTM katman her bir zaman admnda mevcut zaman admndaki veriyi alr ve bir sonraki zaman admnn tahminini yapmak iin bu veriyle birlikte nceki zaman admlarndaki bilgileri kullanr.

Eer return_sequences=True olarak ayarlanrsa, LSTM katman her zaman admndan tam bir zaman serisi kts retir. Yani, her bir zaman admnda bir kt vektr elde edilir ve bu ktlar ardk olarak bir sonraki LSTM katmanna veya kt katmanna geirilebilir. Bu durumda, LSTM katmannn her zaman adm iin bir kt oluturmas ve sonu olarak tam bir zaman serisi kts salamas beklenir.

Bu zellik, birden ok LSTM katmanyla birlikte kullanldnda zaman serisi verilerini daha derin bir ekilde analiz etmek ve daha karmak ilikileri yakalamak iin kullanlabilir. Ayrca, bu ktlar bir kt katmanna dorudan balayarak tahminlemeler yapmak da mmkndr.

zetlemek gerekirse, return_sequences=True ayar, LSTM katmannn her zaman admndan tam bir zaman serisi kts retmesini salar, bylece ardk zaman admlarndaki ilikileri daha etkili bir ekilde modelleyebiliriz.


Eer return_sequences=False olarak ayarlanrsa, LSTM katman sadece son zaman admnda bir kt retir. Yani, LSTM katmannn yalnzca en son zaman admndan gelen kty dndrmesi beklenir.

Bu durumda, LSTM katman sadece son zaman admnn zetini salar. nceki zaman admlarndaki bilgilerin korunmas veya kullanlmas yerine, sadece en son zaman admndaki bilgileri kullanarak tahminler yaplr.

Bunun pratikteki etkisi, ardk verilerdeki uzun vadeli bamllklar ve ilikileri modellemekte daha zorlanabileceidir. zellikle, zaman serilerindeki dinamik desenleri veya uzun vadeli trendleri yakalamak iin LSTM'nin nceki zaman admlarndaki bilgilere erimesi ve bunlar kullanmas genellikle faydaldr.

Ancak baz durumlarda, yalnzca son zaman admnn kts yeterli olabilir. rnein, zaman serisindeki son zaman admnn tek bana tahminlenmesi gereken bir senaryo veya LSTM katmannn bir baka katmana balanaca ve daha karmak bir modelin paras olaca durumlarda return_sequences=False kullanlabilir.

zetle, return_sequences=False ayar, LSTM katmannn yalnzca son zaman admnda bir kt retmesini salar ve nceki zaman admlarndaki bilgilere eriimini kstlar. Bu durumda, zaman serisi analizinde uzun vadeli bamllklar ve dinamik desenleri modellemekte snrlamalar olabilir.

# EarlyStopping and Validation Generator

```python
early_stop = EarlyStopping(monitor = 'val_loss', patience = 5, restore_best_weights = True)
```

restore_best_weights=True parametresi, EarlyStopping ilemi sonunda en iyi performans gsteren model arlklarnn geri yklenmesini salar.

EarlyStopping, modelin eitimini durdurmak iin kullanlan bir gerileme durdurma yntemidir. Belirli bir metrii (genellikle dorulama kayb) izleyerek, bu metriin belirli bir sre boyunca iyilemedii durumlarda eitimi otomatik olarak sonlandrr. Bu, ar uyum (overfitting) durumlarn nlemek ve modelin genellemesini artrmak iin kullanlan bir stratejidir.

restore_best_weights=True ayar, EarlyStopping ilemi sonunda en iyi performans gsteren model arlklarnn geri yklenmesini salar. Yani, EarlyStopping durduunda, en dk dorulama kaybna sahip olan arlklar kullanlarak modelin orijinal arlklar geri yklenir.

Bu zellik, eitim srecinin erken durmasna ramen, en iyi performans gsteren modelin arlklarnn korunmasn salar. Bu ekilde, modelin en iyi sonular verdii noktadaki arlklar kullanlarak daha sonra tahminlemeler veya deerlendirmeler yaplabilir.

Genel olarak, restore_best_weights=True kullanmak, EarlyStopping ile eitilen modellerde en iyi performans elde etmek iin tavsiye edilen bir uygulamadr.

```python
validation_generator = TimeseriesGenerator(test_scaled, test_scaled, length = length, batch_size = batch_size)
```

```python
model.fit(generator,
            epochs = 20,
            validation_data = validation_generator,
            callbacks = [early_stop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

### Evaluation on Test Data

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def eval_metrics(actual, pred):
    rmse = np.sqrt(mean_squared_error(actual, pred))
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    return print("\nmae:", mae, "\nmse:",mse, "\nrmse:",rmse)
```

```python
predictions_scaled = []

first_eval_batch = train_scaled[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead
    current_pred = model.predict(current_batch)
    
    # store prediction
    predictions_scaled.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis=1)
```

Bu kod blou, bir modelin eitimi tamamlandktan sonra, tahminler yapmak iin kullanlan bir dng iermektedir. Aada adm adm kodun ne yapt anlatlmtr:

predictions_scaled adnda bir bo liste tanmlanr. Bu liste, leklendirilmi tahmin deerlerini depolamak iin kullanlacak.

first_eval_batch, eitim veri setinin son length adet esini ieren bir dilimleme ilemi kullanlarak oluturulur. Bu, tahminlerin balang noktasn belirlemek iin kullanlacaktr. train_scaled veri setinin son length satr, modelin tahmin yapmaya balamak iin kulland ilk girdi olarak alnr.

current_batch deikeni, first_eval_batch zerinde bir ekle dntrlr. Bu, tahmin yapmak iin kullanlacak olan mevcut girdiyi temsil eder. Bu dngnn her admnda gncellenecektir.

Bir dng balatlr ve bu dng length adm kadar devam eder.

erideki kod blou, modelden tahminler yapmak iin mevcut girdi zerinde model.predict() ilevini kullanr. current_batch zerinde tahmin yaplr ve current_pred deikenine atlr. Bu tahmin, leklendirilmi bir deerdir.

current_pred deeri, predictions_scaled listesine eklenir. Bylece, her admda yaplan tahminler bu listede saklanr.

current_batch gncellenir. Tahmin deeri, current_batch iine eklenir ve ilk deeri karlr. Bu ekilde, tahminlerin ardk olarak gereklemesi ve yeni tahminlerin girdi olarak kullanlmas salanr.

Bu kod blou, modelin son eitim verilerine dayanarak bir ngr balatr ve ardk olarak yeni tahminler yapar. Tahminler, predictions_scaled listesinde depolanr ve daha sonra leklendirilmi halleri kullanlarak geri dnme tabi tutulabilir. Bu ekilde, modelin ardk tahminler yapma yetenei gsterilmi olur.

```python
current_pred
```

```python
current_batch
```

Grne gre current_batch ve predictions_scaled listesi ilk bata ayn deeri ieriyor. Bunun nedeni, ilk tahminin, modelin balang girdisi olan first_eval_batch zerinde yaplm olmasdr. lk tahmin, current_pred deikenine atldktan sonra predictions_scaled listesine eklenir.

Daha sonra dng, current_batch zerinde tahmin yapmak ve yeni tahmin deerini current_pred olarak gncellemek iin devam eder. Ancak, current_batch gncellenirken, nceki tahmin deeri yeni current_batch'in sonuna eklenirken, ilk deeri atlr. Bu nedenle, her bir dng admnda bir nceki tahmin deeri current_batch iinde yer alr.

Sonu olarak, ilk bata current_batch ve predictions_scaled listesi ayn deeri ierir, ancak dngnn her admnda current_batch gncellenirken yeni tahmin deeri current_pred eklendii iin predictions_scaled listesi tahmin deerlerini saklar.

Bu durum, ardk tahminler yapmann mantn yanstmaktadr. Her bir tahmin, bir sonraki tahmin iin girdi olarak kullanlmak zere current_batch iine eklenir. Bu ekilde, modelin ardk tahminler yapmas salanr.

### Inverse Transformation and Comparing


RNN (Recurrent Neural Network) ile yaplan tahminler genellikle leklendirme ilemi uygulanarak gerek veri leine dntrlr. Bu nedenle, tahminleri gerek lekli deerlere dntrmek iin "Inverse Transformations" (ters dnm) ad verilen bir ilem yaplr.

nceden tahmin edilen deerler predictions_scaled adl bir listede bulunur. Bu tahminler, leklendirme ilemi uygulanm halleridir. Gerek deerlere dntrmek iin, bu tahminlerin leklendirme ilemini geri almak gerekmektedir.

scaler.inverse_transform ilevi, leklendirme ilemini geri alarak tahminlerin gerek lekli deerlerini elde etmek iin kullanlr. Bu ilev, tahminleri ters leklendirme ilemine tabi tutarak gerek deerlere dntrr.

Sonu olarak, predictions_scaled listesindeki tahminlerin gerek lekli deerlerini elde etmek iin scaler.inverse_transform ilevi kullanlr ve sonular predictions adl yeni bir deikende saklanr. Bu sayede, tahminler gerek veri leinde karlatrma ve deerlendirme yapmak iin kullanlabilir.

```python
predictions = scaler.inverse_transform(predictions_scaled)
```

```python
compare = test[:12]
compare['Predictions_tanh'] = predictions
compare
```

```python
eval_metrics(compare.Production, compare.Predictions_tanh)
```

# Retrain and Forecasting - Create Final Model with Full Data

```python
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length = length, batch_size = batch_size)
```

```python
model = Sequential()

model.add(LSTM(64, activation = 'relu', return_sequences=True, input_shape = (length, n_features)))
model.add(LSTM(32, activation = 'relu', return_sequences=True))
model.add(LSTM(16, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')

model.fit(generator, epochs=20)
```

```python
forecast = []

first_eval_batch = scaled_full_data[-length:]
current_batch = first_eval_batch.reshape((1, length, n_features))

for i in range(length):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model.predict(current_batch)
    
    # store prediction
    forecast.append(current_pred[0]) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:, 1:, :], [current_pred], axis = 1)
```

```python
forecast = scaler.inverse_transform(forecast)
forecast
```
